## Print Directory Tree

```python
from pathlib import Path
from toolbox import tree, glob_multiple_patterns
```

### Exclude some folders

```python
for line in tree(Path('.')):
    print(line)
```

```python
list(Path('.').glob('**/*'))
```

```python
for line in tree(Path('.'), exclude_glob=['.ipynb_checkpoints', '**/.git', '**/__pycache__']):
    print(line)
```

### Exclude folders, include only certain extensions

```python
for line in tree(Path('.'), include_glob=['*.ipynb', '*.csv'], exclude_glob=['.venv']):
    print(line)
```
